```python
### 作业
import re
from datetime import datetime,timezone,timedelta
def to_timestamp(dt_str,tz_str):
    m = re.match(r'^UTC([+|-])(0?[0-9]|1[0-2])\:00$',tz_str)
    s = int(m.group(1) + m.group(2))
    dt_dt = datetime.strptime(dt_str,'%Y-%m-%d %H:%M:%S')
    dt_utc_7 = timezone(timedelta(hours=s))
    dt = dt_dt.replace(tzinfo=dt_utc_7)
    return dt.timestamp()
to_timestamp('2015-6-1 08:10:30','UTC+7:00')
datetime.strptime('2015-6-1 08:10:30','%Y-%m-%d %H:%M:%S').timestamp()
```

```python
##内建模块 collections
#nametuple 具备tuple的不变性 还可以定义数据类型 并使用属性来引用
from collections import namedtuple
Point = namedtuple('Point',['x','y'])
p = Point(2,3)
isinstance(p,tuple) #True  Point是Tuple得一种子类
#deque 双向队列  提高插入删除效率  
from collections import deque
q = deque(['a','b','c'])
q.append('x')
q.appendleft('y')
#defalutdict  key不存在是返回默认值
from collections import defaultdict
dd = defaultdict(lambda:'N/A')
dd['key1'] = 'abc'
dd['key2']
#OrderedDict
from collections import OrderedDict
d = dict([('a',1),('b',2),('c',3),('d',4),('e',2),('f',3)])
od = OrderedDict([('a',1),('b',2),('c',3),('d',4),('e',2),('f',3)])
#使用OrderedDict实现先进先出的Dict
class LastUpdatedOrderedDict(OrderedDict):
    def __init__(self,capacity):
        super(LastUpdatedOrderedDict,self).__init__()
        self._capacity = capacity
    
    def __setitem__(self,key,value):
        containsKey = 1 if key in self else 0
        if len(self) - containsKey >= self._capacity:
            last = self.popitem(last=False)
            print('remove',last)
        if containsKey:
#             self[key] = value
            print('set',(key,value))
        else:
            print('add',(key,value))
        OrderedDict.__setitem__(self,key,value)
lo = LastUpdatedOrderedDict(3)
lo['a'] = 1
lo['b'] = 2
lo['c'] = 3
lo['b'] = 4
lo['d'] = 5
print(lo.items())
#Counter计数器
from collections import Counter
c = Counter()
for ch in 'NikoBelicIsAGoodBoy':
    c[ch] = c[ch] + 1
```

```python
#struct
import io,struct
with open("C://Users//lxw//Desktop//test.bmp","rb") as f:
    s = f.read(30)
    print(struct.unpack('<ccIIIIIIHH',s))
def check_bmp(filename):
    with open(filename,"rb") as f:
        t = struct.unpack('<ccIIIIIIHH',s)
        if (t[0] == b'B' and t[1] == b'M'):
            print("Size:",t[2],",Colors:",t[-1])
check_bmp("C://Users//lxw//Desktop//test.bmp")
```

```python
#hashlib
import hashlib
md5 = hashlib.md5()
md5.update("NikoBelic is learning python!!".encode('utf-8'))
print("MD5:",md5.hexdigest())
sha1 = hashlib.sha1()
sha1.update("How to learn python".encode("utf-8"))
sha1.update("enen".encode("utf-8"))
print("SHA1:",sha1.hexdigest())
```

```python
#MD5登陆练习
def calc_md5(password):
    md5 = hashlib.md5()
    md5.update(password.encode('utf-8'))
    return md5.hexdigest()
def login(user,password):
    db = {
        'niko':'e10adc3949ba59abbe56e057f20f883e',
        'tom':'21218cca77804d2ba1922c33e0151105',
        'jerry':'5f4dcc3b5aa765d61d8327deb882cf99'
         }
    try:
        if db[user] == calc_md5(password):
            print("Login Successful!")
        else:
            print("Password Error")
    except:
        print("User Error")
if __name__ == '__main__':
    login('tom','888888')
```

```python
#itertools
import itertools 
natuals = itertools.count(1)
for x in natuals:
    print(x)
    if (x > 10):
        break
        
cs = itertools.cycle('ABC')
i = 0
for x in cs:
    i = i+1
    print(x)
    if i > 10:
        break
        
ns = itertools.repeat('A',3)
for x in ns:
    print(x)

naturals = itertools.count(1)
ns = itertools.takewhile(lambda x : x<10,naturals)
list(ns)
i = 0
print('====================')
for c in itertools.chain('ABC','DEF'):
    print(c)
    i = i + 1
    if i > 20:
        break

for key,group in itertools.groupby('AaaBbbCCc',lambda c : c.upper()):
    print(key,list(group))

```

```python
#XML
from xml.parsers.expat import ParserCreate

class DefaultSaxHandler(object):
    def start_element(self,name,attrs):
        print('SAX start_element: %s ,attrs: %s' % (name,attrs))
    def end_element(self,name):
        print("SAX end_element: %s" % name)
    def char_data(self,text):
        print('SAX char_data: %s' % text)
xml = r'''<?xml version="1.0"?>
<ol>
    <li><a href="/python">Python</a></li>
    <li><a href="/ruby">Ruby</a></li>
</ol>
'''
handler  = DefaultSaxHandler()
parser = ParserCreate()
parser.StartElementHandler = handler.start_element
parser.EndElementHandler = handler.end_element
parser.CharacterDataHandler = handler.char_data
parser.Parse(xml)
```

```python
#POST
#模拟微博登陆
from urllib import request, parse

print('Login to weibo.cn...')
email = input('Email: ')
passwd = input('Password: ')
login_data = parse.urlencode([
    ('username', email),
    ('password', passwd),
    ('entry', 'mweibo'),
    ('client_id', ''),
    ('savestate', '1'),
    ('ec', ''),
    ('pagerefer', 'https://passport.weibo.cn/signin/welcome?entry=mweibo&r=http%3A%2F%2Fm.weibo.cn%2F')
])
print(login_data)
req = request.Request('https://passport.weibo.cn/sso/login')
req.add_header('Origin', 'https://passport.weibo.cn')
req.add_header('User-Agent', 'Mozilla/6.0 (iPhone; CPU iPhone OS 8_0 like Mac OS X) AppleWebKit/536.26 (KHTML, like Gecko) Version/8.0 Mobile/10A5376e Safari/8536.25')
req.add_header('Referer', 'https://passport.weibo.cn/signin/login?entry=mweibo&res=wel&wm=3349&r=http%3A%2F%2Fm.weibo.cn%2F')

with request.urlopen(req, data=login_data.encode('utf-8')) as f:
    print('Status:', f.status, f.reason)
    for k, v in f.getheaders():
        print('%s: %s' % (k, v))
    print('Data:', f.read().decode('utf-8'))
```
